```python
import codecs
f = codecs.open('Nat2018PublicUS.c20190509.r20190717.txt', mode='r', encoding='utf-8')
line = f.readline()
list1 = []
while line:
    a = line.split()
    b = len(a)
    if b == 54:
        list1.append(a[52])
    elif b == 55:
        list1.append(a[53])
    elif b == 53:
        list1.append(a[51])
    else:
        list1.append('')
    line = f.readline()
f.close()
```

```python
list1
```

```python
import pandas as pd
test= pd.DataFrame(data=list1)
print(test)
test.to_csv('DiseaseType.csv',encoding='gbk')
```
