```python
import pandas as pd
```

```python
import numpy as np
```

```python
a = pd.read_csv('C:/Users/a8809/Desktop/Big Data/project/US_births(2018)/US_births(2018).csv', chunksize=100000)
b = pd.read_csv('C:/Users/a8809/Desktop/Big Data/project/DiseaseType.csv', chunksize=100000)
c = pd.read_csv('C:/Users/a8809/Desktop/Big Data/project/APGAR_SCORE.csv', chunksize=100000)
```

```python
df1 = pd.concat([chunk for chunk in a], axis=0)
df2 = pd.concat([chunk for chunk in b], axis=0)
df3 = pd.concat([chunk for chunk in c], axis=0)
```

```python
df = df1.join(df2, lsuffix='_df1', rsuffix='_df2')
```

```python
df = df.join(df3, lsuffix='_df', rsuffix='_df3')
```

```python
del df1
del df2
del df3
```

```python
df = df.drop(columns=['DLMP_YY', 'DOB_YY', 'FRACE15', 'FRACE31', 'IMP_SEX', 'LD_INDL', 'MAR_IMP', 'MRACE15',
                 'MRACE31', 'MRACEIMP', 'PAY', 'RF_CESAR', 'Unnamed: 0_df','Unnamed: 0.1', '0_df', 'Unnamed: 0_df3'])
```
