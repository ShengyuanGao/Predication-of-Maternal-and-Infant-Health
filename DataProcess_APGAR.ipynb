```python
import pandas as pd
```

```python
import codecs
f = codecs.open('Nat2018PublicUS.c20190509.r20190717.txt', mode='r', encoding='utf-8')
line = f.readline()
list1 = []
while line:
    a= line.split()
    b=  a[40]
    c=  a[39]
    d = a[41]
    if len(b) == 7:
        list1.append(b[0:2])
    elif len(b) == 6:
        list1.append(d[0:2])
    elif len(b) == 1:
        list1.append(c[0:2])
    line = f.readline()
f.close()
```

```python
test= pd.DataFrame(data=list1)
print(test)
test.to_csv('APGAR_SCORE.csv',encoding='gbk')
```
